## ex. 2

$$ y = tr(AB) \quad A,B \in \mathbb{R}^{N \times N} $$

$$
\frac{dy}{dA} =
$$

$$
\frac{dy}{dA} = B^T
$$

**Вывод:**

След — линейный оператор, поэтому дифференциал можно занести внутрь.
$$
dy = d(\text{tr}(AB)) = \text{tr}(d(AB))
$$


По правилу произведения ( $B$ константа) и циклическому свойству следа:
$$
\text{tr}(d(AB)) = \text{tr}((dA)B) = \text{tr}(B(dA))
$$

Сравним  $dy = \text{tr}(B \cdot dA)$ с эталонной формой     $$ dy = \left\langle \frac{dy}{dA}, dA \right\rangle = \text{tr}\left( \left(\frac{dy}{dA}\right)^T dA \right) $$, получаем:
$$
\left(\frac{dy}{dA}\right)^T = B
$$
Откуда, транспонируя :
$$
\frac{dy}{dA} = B^T
$$

### Question 2

We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$,

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply.
1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data.

*Your Answer:*
**1, 2, 3**

*Your Explanation:*
Результат работы k-NN зависит только от порядка соседей. Если преобразование не меняет порядок (кто был ближе, тот и остался ближе), результат не изменится.
1.  **Вычитание общего среднего:** Взаимные расстояния не меняются.
2.  **Вычитание попиксельного среднего:** Это сдвиг вдоль каждой оси на свою константу. При вычислении разности эти сдвиги сокращаются, и расстояние не меняется.
3.  **Стандартизация с общими параметрами:** Все расстояния умножаются на одну и ту же константу $1/\sigma$. Это масштабирует все расстояния одинаково, не меняя порядка ближайших соседей.
4.  *Попиксельная стандартизация* меняет результат, так как это эквивалентно введению весов для каждой координаты, что изменяет метрику расстояния.
5.  *Поворот* меняет результат, так как L1  не инвариантно к вращениям.
